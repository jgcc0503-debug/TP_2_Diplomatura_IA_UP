```python
# ==========================================
# 02_preprocesamiento.ipynb
# ==========================================
import pandas as pd
import os
from google.colab import drive
from sklearn.preprocessing import OrdinalEncoder
```

```python
# 1. Cargar datos del dataset de entrenamiento
drive.mount('/content/drive')
ruta_train = '/content/drive/My Drive/Training_And_Testing_Dataset/smoking_prediction.csv'
df_train = pd.read_csv(ruta_train)
# 2. Identificar columnas categóricas (las que tienen texto)
columnas_categoricas = df_train.select_dtypes(include=['object']).columns.tolist()

# 3. Transformación (Conversión de texto a números)
# Usamos OrdinalEncoder para convertir cada categoría en un número (ej: Male=0, Female=1)
encoder = OrdinalEncoder()
df_train[columnas_categoricas] = encoder.fit_transform(df_train[columnas_categoricas])

# 4. Se eliminan datos constantes que no aporten a la identificación de fumadores
cols_a_eliminar = [col for col in df_train.columns if df_train[col].nunique() <= 1]

if cols_a_eliminar:
    print(f"Eliminando columnas constantes que no aportan información: {cols_a_eliminar}")
    df_train = df_train.drop(columns=cols_a_eliminar)#

# 5. Guardar en la carpeta procesada
ruta_processed = '/content/drive/My Drive/Training_And_Testing_Dataset/smoking_prediction_processed.csv'
df_train.to_csv(ruta_processed, index=False)

print(f"✅ Datos procesados y guardados en: {ruta_processed}")
```
